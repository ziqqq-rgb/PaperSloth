```python
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import (
    PdfPipelineOptions,
    AcceleratorOptions,
    AcceleratorDevice,
)

pipeline_options = PdfPipelineOptions()
pipeline_options.accelerator_options = AcceleratorOptions(
    num_threads=4,
    device=AcceleratorDevice.CPU  
)

converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
    }
)

result = converter.convert("./test1.pdf")
doc = result.document
print(doc)
```

```python
import json
from llama_index.core import Document
from llama_index.node_parser.docling import DoclingNodeParser
from docling_core.transforms.chunker.hybrid_chunker import HybridChunker

doc_json_str = json.dumps(doc.export_to_dict())

# Wrap it into a LlamaIndex Document object
llama_doc = Document(text=doc_json_str)

chunker = HybridChunker(
    tokenizer="nomic-ai/nomic-embed-text-v2-moe",
    max_tokens=512
)

node_parser = DoclingNodeParser(chunker=chunker)

# Pass the manually wrapped document
nodes = node_parser.get_nodes_from_documents([llama_doc])
print(nodes)
```

```python
# Print structural element counts from the DoclingDocument
print("=== Document Elements Summary ===")
print(f"Total Pages:  {len(doc.pages)}")
print(f"Total Tables: {len(doc.tables)}")
print(f"Total Images: {len(doc.pictures)}")
print(f"Total Texts:  {len(doc.texts)}")

# Optional breakdown: Let's see what kinds of text elements are inside
from docling_core.types.doc.document import GroupItem

print("\n=== Text Type Breakdown ===")
text_types = {}
for item in doc.texts:
    # Safely extract the structural type/label of each text segment
    label = getattr(item, "label", "unknown")
    text_types[label] = text_types.get(label, 0) + 1

for label, count in sorted(text_types.items(), key=lambda x: x[1], reverse=True):
    print(f"- {label.capitalize()}: {count}")

```
